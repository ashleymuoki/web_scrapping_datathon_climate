```python
#relationship analysis
#1 correlation matrix
correlation = projects.corr()
```

```python
sns.heatmap(correlation, xticklabels= correlation.columns,yticklabels=correlation.columns,annot=True)
```

```python
#use pairplot
sns.pairplot(projects, )
```

```python
#scatter plot 1
sns.relplot(x='Total Project Cost $US',y = 'Total IDA and IBRD Commitment $US', hue= 'Region',data = projects)
```

```python
#scatter plot 2
sns.relplot(x='Total Project Cost $US',y = 'Total IDA and IBRD Commitment $US', hue= 'Environmental Assessment Category',data = projects)
```

```python
#scatter plot 3
sns.relplot(x='Total Project Cost $US',y = 'Total IDA and IBRD Commitment $US', hue= 'Project Status',data = projects)
```

```python
#bar chart 1
x = projects['Region']
y = projects['Total IDA and IBRD Commitment $US']
plt.figure(figsize=(5,5))
plt.title("Region Against Total IDA and IBRD Commitment $US", fontsize= 20)
plt.xlabel('Total IDA and IBRD Commitment $US', fontsize = 16, rotation=45)
plt.ylabel('Region', fontsize=16)
plt.barh(x,y)
```

```python
#bar chart 2
x2 = projects['Region']
y2 = projects['Total Project Cost $US']
plt.figure(figsize=(5,5))
plt.title("Region Against Total Project Cost $US", fontsize= 20)
plt.xlabel('Total Project Cost $US', fontsize = 16, rotation=45)
plt.ylabel('Region', fontsize=16)
plt.barh(x2,y2)
```

```python
#bar chart 3
x3 = projects['Project Status']
y3 = projects['Total IDA and IBRD Commitment $US']
plt.figure(figsize=(5,5))
plt.title("Project status Against Total IDA and IBRD Commitment $US", fontsize= 20)
plt.xlabel('Total IDA and IBRD Commitment $US', fontsize = 16, rotation=45)
plt.ylabel('Project Status', fontsize=16)
plt.barh(x3,y3)
```

```python
#bar chart 4
x4 = projects['Project Status']
y4 = projects['Total Project Cost $US']
plt.figure(figsize=(5,5))
plt.title("Project Status Against Total Project Cost $US", fontsize= 20)
plt.xlabel('Total Project Cost $US', fontsize = 16, rotation=45)
plt.ylabel('Project Status', fontsize=16)
plt.barh(x4,y4)
```

```python
#histogram 1
sns.histplot(x = 'Total Project Cost $US', data = projects, bins = 5)
```
